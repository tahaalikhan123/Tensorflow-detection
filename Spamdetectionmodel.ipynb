```python
#import dependencies
import pandas as pd
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
#Load your dataset
df = pd.read_csv('spamdataset.csv')
```

```python
#Explore the dataset
print(df.head())
```

```python
#preprocessing the dataset
#here we split the dataset into two sets, training and testing
train_texts, test_texts, train_labels, test_labels = train_test_split(df['Message'], df['Category'], test_size=0.2, random_state=42)
```

```python
#tokenize text and pad sequences in the text data
tokenizer = Tokenizer()
tokenizer.fit_on_texts(train_texts)

train_sequences = tokenizer.texts_to_sequences(train_texts)
test_sequences = tokenizer.texts_to_sequences(test_texts)

max_len = 100  # Set the maximum sequence length
train_data = pad_sequences(train_sequences, maxlen=max_len)
test_data = pad_sequences(test_sequences, maxlen=max_len)
```

```python
#let's build the model
model = Sequential()
model.add(Dense(128, activation='relu', input_shape=(max_len,)))
model.add(Dense(1, activation='sigmoid'))

# Use the recommended loss function for binary classification
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
#rerun data preprocessing
# Split the dataset into training and testing sets
train_texts, test_texts, train_labels, test_labels = train_test_split(df['Message'], df['Category'], test_size=0.2, random_state=42)

# Tokenize and pad the text data
tokenizer = Tokenizer()
tokenizer.fit_on_texts(train_texts)

train_sequences = tokenizer.texts_to_sequences(train_texts)
test_sequences = tokenizer.texts_to_sequences(test_texts)

max_len = 100  # Set the maximum sequence length
train_data = pad_sequences(train_sequences, maxlen=max_len)
test_data = pad_sequences(test_sequences, maxlen=max_len)
```

```python
#rebuild the model
model = Sequential()
model.add(Dense(128, activation='relu', input_shape=(max_len,)))
model.add(Dense(1, activation='sigmoid'))

# Use binary_crossentropy loss for binary classification
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Train the Model
model.fit(train_data, train_labels, epochs=5, batch_size=32, validation_split=0.2)
```

```python
#Evaluate the Model
# Evaluate the model on the testing data
loss, accuracy = model.evaluate(test_data, test_labels)
print(f'Test Loss: {loss:.4f}, Test Accuracy: {accuracy:.4f}')
```

```python
#Visualizing Training History:
import matplotlib.pyplot as plt

# Train the model and store the training history
history = model.fit(train_data, train_labels, epochs=5, batch_size=32, validation_split=0.2)

# Plot training loss and validation loss over epochs
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

```python
#visualizing confusion matrix
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

# Make predictions on the test set
predictions = model.predict(test_data)

# Convert probabilities to binary labels based on a threshold
threshold = 0.5
predicted_labels = (predictions > threshold).astype(int)

# Create a confusion matrix
cm = confusion_matrix(test_labels, predicted_labels)

# Plot the confusion matrix using seaborn
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['Non-Spam', 'Spam'], yticklabels=['Non-Spam', 'Spam'])
plt.xlabel('Predicted')
plt.ylabel('True')
plt.show()
```
